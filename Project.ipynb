### 1 Summarise the history of the box plot and situations in which it used

***Box plot is a histogram-like method of displaying data, invented by J. Tukey***

- Box Plot

Boxplot is a method for graphically depicting groups of numerical data through their quartiles


   


***Median***

The median  marks the midpoint of the data and is shown by the line that divides the box into two parts

***Inter-quartile range***

The middle box represents the middle 50% of scores for the group.

***Upper quartile***

75 % of the scores fall below the upper quartile.

***Lower quartile***

25 % of scores fall below the lower quartile.

***Whiskers***

The upper and lower whiskers represent scores outside the middle 50%.

***Box Plot History***

- The basic graphic form of the boxplot  was in early 1950

- The first variation to display a density estimate was the vase plot(Benjamini, 1988), where the box is replaced with a symmetrical display of estimated density.

- (Hyndman and Fan, 1996)boxplots replace the extremes with fixed quantiles(min and max,2% and 98%)

- The highest density region(HDR)boxplot(Hyndman,1996)is compromise between a boxplot and a density boxplot

- Hintze and Nelson In 1998 vase plot was replaced Violin plots this was very similar, but display the density for all data points

- The bagplot (Rousseuw 1999)is a 2nd analog of the boxplot  with  a bag and fence box containing the middle 50% of the data and a loop to separate outliers

- The rotational boxplot was introduce in 2000
 
- In an effort to improve the data ink ratio of the boxplot(Tufte, 2001)
 
- The box percentile plot (Esty and Banfield, 2003)displays a modified empirical cumulative density function 

- The letter value boxplot(Hofmann 2006)was  designed to overcome the shortcomings of the boxplot for large data.

- Skewness and multimodalityand mean andstandard error(Marmolejo-Ramos and Tian,2010)



***Why box plots***

- Handles Large Data Easily

   A box plot can handle and present a summary of a large amount of data. it consists of the median, which is the midpoint of the range of data the upper and lower quartiles, which represent the numbers above and below the highest and lower quarters of the data and the minimum and maximum data values.
   
    
- Clear summary

    A box plot is a highly visually effective way of viewing a clear summary of one or more sets of data. It is particularly useful for quickly summarizing and comparing different sets of results from different experiments.
    
    
- Displays outliers

    A box plot is one of very few statistical graph methods that show outliers. There might be one outlier or multiple outliers within a set of data, which occurs both below and above the minimum and maximum data values. By extending the lesser and greater data values to a max of 1.5 times the inter-quartile range, the box plot delivers outliers or obscure results. Any results of data that fall outside of the minimum and maximum values known as outliers are easy to determine on a box plot graph.


***When it Used ***

- It is often used in explanatory data analysis. 

- Graph is used to show the shape of the distribution.

- The ends of the box are the upper and lower quartiles, so the box spans the interquartile range.

### 4. Compare the box plot against other plots

Box plots are most suited to displaying the distribution of a variable across multiple groups.


    Going to compare box plot to other types of plot using seaborn library
    Create Box plot ,Violin plot,Scattor Plot and swarm plot for the same dataset 
    Observations of the different plots


```python
import matplotlib.pyplot as plt
#Creates two subplots and unpacks the output array immediately
f, axes = plt.subplots(1, 4, figsize=(15, 5))
#Draw a box plot to show distributions with respect to categories.
sns.boxplot(x="day", y="tip", data=tips, ax=axes[0])
#Draw a scatter plot with possibility of several semantic groupings.
sns.scatterplot(x="day", y="tip", hue="day", data=tips, ax=axes[1])
sns.swarmplot(x="day", y="tip", data=tips, ax=axes[2])
sns.violinplot(x="day", y="tip", data=tips, palette="muted",inner="quartile",  scale="count", ax=axes[3])
```
