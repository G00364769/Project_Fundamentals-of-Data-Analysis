***Box plot is a histogram-like method of displaying data, invented by J. Tukey***

***Box Plot***

Boxplot is a method for graphically depicting groups of numerical data through their quartiles. Box plots may also have lines extending vertically from the boxes (whiskers) indicating variability outside the upper and lower quartiles

A simple way of representing statistical data on a plot in which a rectangle is drawn to represent the second and third quartiles, usually with a vertical line inside to indicate the median value. The lower and upper quartiles are shown as horizontal lines either side of the rectangle.




***Median***

The median (middle quartile) marks the mid-point of the data and is shown by the line that divides the box into two parts. Half the scores are greater than or equal to this value and half are less.

***Inter-quartile range***

The middle “box” represents the middle 50% of scores for the group. The range of scores from lower to upper quartile is referred to as the inter-quartile range. The middle 50% of scores fall within the inter-quartile range.

***Upper quartile***

Seventy-five percent of the scores fall below the upper quartile.

***Lower quartile***

Twenty-five percent of scores fall below the lower quartile.

***Whiskers***

The upper and lower whiskers represent scores outside the middle 50%. Whiskers often (but not always) stretch over a wider range of scores than the middle quartile groups.

***Box Plot History***

- The basic graphic form of the boxplot, the range-bar, was established in the early 1950’

- The first variation to display a density estimate was the vase plot(Benjamini, 1988), where the box is replaced with a symmetrical display of estimated density.

- (Hyndman and Fan, 1996), some boxplots replace the extremes with fixed quantiles (e.g. min and max, 2% and 98%)

- The highest density region (HDR) boxplot (Hyndman, 1996) is a compromise between a boxplot and a density boxplot

- Hintze and Nelson In 1998 vase plot was replaced Violin plots this was very similar, but display the density for all data points, not just the middle half

- The bagplot (Rousseuw 1999)is a 2d analog  of  the  boxplot  with  a bag and fence (box) containing the middle 50% of the data and a loop (whiskers)to separate outliers

- The rotational boxplot was introduce in (2000)
 
- In an effort to improve the data-ink ratio of the boxplot, (Tufte, 2001)
 
- The box-percentile plot (Esty and Banfield, 2003) displays a modified empirical cumulative density function (ECDF)

- The  letter-value  boxplot  (Hofmann 2006)  was  designed  to  overcome  the  shortcomings  of  the boxplot for large data.

- Skewness and multimodalityand mean andstandard error (Marmolejo-Ramos and Tian, 2010)



***Why box plots***

- Handles Large Data Easily

   Due to the five-number data summary, a box plot can handle and present a summary of a large amount of data. A box plot consists of the median, which is the midpoint of the range of data; the upper and lower quartiles, which represent the numbers above and below the highest and lower quarters of the data and the minimum and maximum data values. Organizing data in a box plot by using five key concepts is an efficient way of dealing with large data too unmanageable for other graphs, such as line plots or stem and leaf plots.
   
    
- Exact Values Not Retained

    The box plot does not keep the exact values and details of the distribution results, which is an issue with handling such large amounts of data in this graph type. A box plot shows only a simple summary of the distribution of results so that you can quickly view it and compare it with other data. Use a box plot in combination with another statistical graph method, like a histogram, for a more thorough, more detailed analysis of the data.
    
- A clear summary

    A box plot is a highly visually effective way of viewing a clear summary of one or more sets of data. It is particularly useful for quickly summarizing and comparing different sets of results from different experiments. At a glance, a box plot allows a graphical display of the distribution of results and provides indications of symmetry within the data.
    
    
- Displays outliers

    A box plot is one of very few statistical graph methods that show outliers. There might be one outlier or multiple outliers within a set of data, which occurs both below and above the minimum and maximum data values. By extending the lesser and greater data values to a max of 1.5 times the inter-quartile range, the box plot delivers outliers or obscure results. Any results of data that fall outside of the minimum and maximum values known as outliers are easy to determine on a box plot graph.


***When it Used ***

It is often used in explanatory data analysis. This type of graph is used to show the shape of the distribution, its central value, and its variability. In a box and whisker plot: the ends of the box are the upper and lower quartiles, so the box spans the interquartile range.

***Percentiles and Quartiles***

The first quartile (or lower quartile), Q1, is defined as the value that has an f-value equal to 0.25. This is the same thing as the twenty-fifth percentile. The third quartile (or upper quartile), Q3, has an f-value equal to 0.75.

**Median**

The median is the score of a distribution residing at the 50th percentile, separating the top and bottom 50 percent of scores. The median is useful for both splitting a set of distribution scores in half and helping to identify the skew of a distribution.

**Interquartile Range (IQR)**

The IQR is the difference between the score delineating the 75th percentile and the 25th percentile, the third and first quartiles, respectively.

```python
#iris.groupby('species')['sepal_length'].median()

# wring a function that will give percentile , Median , Min,Max,Std
def q1(x):
    return x.quantile(0.25)

def q2(x):
    return x.quantile(0.75)
# percentile , Median , Min,Max,Std
f = ['median', 'std','min','max', q1,q2]
df1 = iris.groupby('species')['sepal_length','sepal_width'].agg(f)

#print(df1)
df1
```

***q1 = equal to 0.25 and q2=equal to 0.75***

```python
print("The medians of Sepal Length" ,iris.groupby('species')['sepal_length'].median())
print ("\n")
print("The medians of Sepal Width" ,iris.groupby('species')['sepal_width'].median())
print ("\n")
print("The quantile(0.25) of Sepal Length " ,iris.groupby('species')['sepal_length'].agg(q1))
print ("\n")
print("The quantile(0.25) of Sepal Width " ,iris.groupby('species')['sepal_width'].agg(q1))
print ("\n")
print("The quantile(0.75) of Sepal Length " ,iris.groupby('species')['sepal_length'].agg(q2))
print ("\n")
print("The quantile(0.75) of Sepal Width " ,iris.groupby('species')['sepal_width'].agg(q2))
print ("\n")
print("Min:",df1.min()) 
```

***Comparing Box Plots and Histograms***

A histogram is a type of bar chart that graphically displays the frequencies of a data set. Similar to a bar chart, a histogram plots the frequency, or raw count, on the Y-axis (vertical) and the variable being measured on the X-axis (horizontal).

A box plot, also called a box-and-whisker plot, is a chart that graphically represents the five most important descriptive values for a data set. These values include the minimum value, the first quartile, the median, the third quartile, and the maximum value. When graphing this five-number summary, only the horizontal axis displays values. 

```python
import seaborn as sns
import matplotlib.pyplot as plt
tips = sns.load_dataset("tips")
tips.head()
```

***Drawing a histogram with a boxplot appearing the top of the histogram showing the Q1,Q2 and Q3 as well as the outliers***

```python
# Import library and dataset
import seaborn as sns
import numpy as np
df = sns.load_dataset('tips')
 
# Partition the plot window to 2 parts
f, (ax_box, ax_hist) = plt.subplots(2, sharex=True, gridspec_kw={"height_ratios": (.12, .35)})

# Add a graph in each window:
sns.boxplot(df["total_bill"], ax=ax_box)
sns.distplot(df["tip"], ax=ax_hist ) 
ax_box.set(yticks=[])
sns.despine(ax=ax_hist)
sns.despine(ax=ax_box, left=True)


print("For  BOX plot Min is ",tips['total_bill'].min())
print ("\n")
print(" max is ",tips['total_bill'].max())
print ("\n")
print(" Median is ",tips['total_bill'].median())
print ("\n")
print(" Q1(25) is ",np.percentile(tips['total_bill'], 25))
print ("\n")
print("Q2(75) is ",np.percentile(tips['total_bill'], 75))
```

Box Plot

Box plots are most suited to displaying the distribution of a variable across multiple groups. The bottom and top of the boxes indicate the lower and upper quartiles, respectively, and the line inside the box is for the median. Vertical lines extending from the boxes ("whiskers") show the range of the data (by default, this is 1.5x past the upper and lower quartiles in matplotlib). Box plots can be thought of as a hybrid between bar plots and overlaid histograms. They surface much of the same information as bar plots, but they also expose the variation in the data. However, they do not show the underlying distribution of the data

```python
#axes =  figsize=(6, 4)
sns.boxplot(x="day", y="tip", data=tips)
Tip= sns.catplot(x="total_bill", y="day", hue="time", height=3.5, aspect=2.5,
                kind="box", legend=False, data=tips);
Tip.add_legend(title="Meal")
Tip.set_axis_labels("Total bill ($)", "")
Tip.set(xlim=(0, 80), yticklabels=["Thursday", "Friday", "Saturday", "Sunday","Monday"])
Tip.fig.set_size_inches(9.5, 4.5)
Tip.ax.set_xticks([5, 15, 25, 35, 45, 55], minor=True)
plt.setp(Tip.ax.get_yticklabels(), rotation=30)
```

***Conclusions***

Data visualization should be regarded as a complement to summary statistics — doing one without the other will leave you with an incomplete and potentially misleading view of your data. Just as choosing the right summary statistic for the situation at hand is critical, so is selecting the right plot type for your purposes. When comparing quantitative variables to each other, consider scatter and line plots. If you are interested in how a variable is distributed, turn to histograms and density plots. For comparisons involving categorical variables, bar and box plots are good starting points.
