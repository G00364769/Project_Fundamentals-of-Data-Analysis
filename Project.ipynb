***Box plot is a histogram-like method of displaying data, invented by J. Tukey***

Boxplot is a method for graphically depicting groups of numerical data through their quartiles. Box plots may also have lines extending vertically from the boxes (whiskers) indicating variability outside the upper and lower quartiles

A simple way of representing statistical data on a plot in which a rectangle is drawn to represent the second and third quartiles, usually with a vertical line inside to indicate the median value. The lower and upper quartiles are shown as horizontal lines either side of the rectangle.



***Median***

The median (middle quartile) marks the mid-point of the data and is shown by the line that divides the box into two parts. Half the scores are greater than or equal to this value and half are less.

***Inter-quartile range***

The middle “box” represents the middle 50% of scores for the group. The range of scores from lower to upper quartile is referred to as the inter-quartile range. The middle 50% of scores fall within the inter-quartile range.

***Upper quartile***

Seventy-five percent of the scores fall below the upper quartile.

***Lower quartile***

Twenty-five percent of scores fall below the lower quartile.

***Whiskers***

The upper and lower whiskers represent scores outside the middle 50%. Whiskers often (but not always) stretch over a wider range of scores than the middle quartile groups.

***When it Used ***

It is often used in explanatory data analysis. This type of graph is used to show the shape of the distribution, its central value, and its variability. In a box and whisker plot: the ends of the box are the upper and lower quartiles, so the box spans the interquartile range.

Using NumPy and matplotlib lib i am going to generate 4 types of box plot. For generate number set i am using random.standard_gamma ,random.exponential and np.random.normal, Random.normal has two variable mean and Standard deviation ,Random.exponential has scale and random.standard_gamma has parameter and size is constant for all four set as N
 

***Types of Box Plots***

```python
import matplotlib.pyplot as plt
import numpy as np
#same set of numbers will appear every time
np.random.seed(123)
#Generating sample data
N = 112
Set1 = np.random.normal(21, 5, N)#Mean=21,Standard deviation=5
Set2 = np.random.normal(6, 2, N)
Set3 = np.random.exponential(4, N)#scale=4, size=N)
Set4 = np.random.standard_gamma(4, N)#Parameter=4


data = [Set1, Set2, Set3, Set4]
fig=plt.figure(figsize=(9,6))
# Adding subplot
ax = fig.add_subplot(111)
# notched plot
NP = ax.boxplot(data,1)
# defining  x cordinate as box list
plt.xticks([1, 2, 3, 4],['BOX1', 'BOX2', 'BOX3', 'BOX4'])
plt.title('Notched plot')
plt.show()
# change outlier point symbols
ct=plt.boxplot(data, 0, 'gD')
plt.xticks([1, 2, 3, 4],['BOX1', 'BOX2', 'BOX3', 'BOX4'])
plt.title('Change outlier point symbols')
plt.show()
# don't show outlier points
plt.boxplot(data, 0, '')
plt.xticks([1, 2, 3, 4],['BOX1', 'BOX2', 'BOX3', 'BOX4'])
plt.title('dont show outlier points')
plt.show()
# horizontal boxes
plt.boxplot(data, 0, 'rs', 0)
plt.yticks([1, 2, 3, 4],['BOX1', 'BOX2', 'BOX3', 'BOX4'])
plt.title('Horizontal boxes')
plt.show()

```
